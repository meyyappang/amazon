```python
import pandas as pd
import selenium
from selenium import webdriver 
from selenium.webdriver.support.ui import Select
import time
from itertools import zip_longest
driverPath=r"C:\Users\Merit\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(executable_path = driverPath)
driver.maximize_window()
driver.get("https://www.amazon.in/s?k=offers+on+laptops&i=computers&crid=XQIAUT3DDH91&sprefix=offers+%2Ccomputers%2C302&ref=nb_sb_ss_ts-doa-p_1_7")

writer = pd.ExcelWriter('Amazon.xlsx',engine = 'xlsxwriter')

lipstick = driver.find_elements_by_xpath("//h2[contains(@class,'s-line-clamp-4')]")
old_price = driver.find_elements_by_xpath("//span[contains(@class,'a-text-price')]")
offer_price = driver.find_elements_by_xpath("//span[contains(@class,'a-price-whole')]")

name = []
oldprice= []
offerprice= []
for i in lipstick:
    name.append(i.text)

for j in old_price:
    oldprice.append(j.text)

for k in offer_price:
    offerprice.append(k.text)

l=zip_longest(name,oldprice,offerprice)

df = pd.DataFrame(l,columns=["Name","old_price","offer_price"])
df.to_excel(writer,sheet_name = 'laptop' ,index = False)
writer.save()
df
```
